### **Завдання 1:** Виконати теоретичні розрахунки асимптотичної складності алгоритмів сортування методами вставляння та бульбашки.
Обидва алгоритми (сортування вставлянням та бульбашкою) базуються на вкладених циклах, що визначає їхню часову складність.

**Сортування бульбашкою** *(Bubble Sort)*

Найгірший та середній випадки: Масив відсортований у зворотному порядку. Кількість порівнянь: $\frac{n(n-1)}{2}$.
$$T(n) = O(n^2)$$
Найкращий випадок: Масив уже відсортований.
$$T(n) = O(n)$$
**Сортування вставлянням** *(Insertion Sort)*

Найгірший та середній випадки: Елементи знаходяться далеко від своїх позицій.
$$T(n) = O(n^2)$$
Найкращий випадок: Масив уже відсортований (лише внутрішня перевірка без зсувів).
$$T(n) = O(n)$$


### **Завдання 2:** Реалізувати на Python алгоритм сортування методом бульбашки.

```python
import time
import random
import matplotlib.pyplot as plt

# Алгоритм сортування бульбашкою
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
```

### **Завдання 3:** Дослідити реальний час виконання обох алгоритмів на Python і побудувати графіки залежностей *Tinsert(n)* та *Tbuble(n)* для n = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000, 20000, 50000, 10000], згенерувавши відповідні цілочисельні масиви.

```python
n_values = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000] # Скорочено для стабільності
t_bubble = []
t_insert = []

for n in n_values:
    # Генерируємо випадковий масив
    test_arr = [random.randint(0, 100000) for _ in range(n)]
    
    # Вимірювання для Bubble Sort
    arr_copy = test_arr.copy()
    start = time.time()
    bubble_sort(arr_copy)
    t_bubble.append(time.time() - start)
    
    # Вимірювання для Insertion Sort
    arr_copy = test_arr.copy()
    start = time.time()
    insertion_sort(arr_copy)
    t_insert.append(time.time() - start)
    

# Побудова графіка
plt.figure(figsize=(10, 6))
plt.plot(n_values, t_bubble, label='Bubble Sort', marker='o')
plt.plot(n_values, t_insert, label='Insertion Sort', marker='s')
plt.xlabel('Кількість елементів (n)')
plt.ylabel('Час виконання (секунди)')
plt.title('Залежність часу сортування від розміру масиву')
plt.legend()
plt.grid(True)
plt.show()
```

1. Що таке O-нотація і чим вона відрізняється від Ω -нотації?

О-нотація: гарантує, що алгоритм не буде працювати повільніше за цей ліміт. Це "найгірший випадок".

Ω -нотація: гарантує, що алгоритм не буде працювати швидше за цей ліміт. Це "найкращий випадок".
