### **Завдання 1:** Створити бінарне дерево згідно з варіантом, виданим викладачем. Написати процедуру видалення заданої гілки дерева. 


```python
# Задання вузла бінарного дерева (корінь, ліве піддерево, праве піддерево)
def BinaryTree(r):
    return [r, [], []]

# Додавання елемента у ліве піддерево
def insertLeft(root, newBranch):
    t = root.pop(1)
    if len(t) > 1:
        root.insert(1, [newBranch, t, []])
    else:
        root.insert(1, [newBranch, [], []])
    return root

# Додавання елемента у праве піддерево
def insertRight(root, newBranch):
    t = root.pop(2)
    if len(t) > 1:
        root.insert(2, [newBranch, [], t])
    else:
        root.insert(2, [newBranch, [], []])
    return root

# Процедура видалення заданої гілки дерева разом із її нащадками
def deleteBranch(root, val):
    if not root:
        return
    
    # Перевіряємо, чи є лівий нащадок шуканою гілкою
    if root[1] and root[1][0] == val:
        root[1] = []  # Повністю очищуємо ліву гілку
        return

    # Перевіряємо, чи є правий нащадок шуканою гілкою
    if root[2] and root[2][0] == val:
        root[2] = []  # Повністю очищуємо праву гілку
        return

    deleteBranch(root[1], val)
    deleteBranch(root[2], val)



# Створюємо початкове дерево
my_tree = BinaryTree(50)
insertLeft(my_tree, 30)
insertRight(my_tree, 70)

# Додаємо нащадків до вузла 30 (ліва гілка)
insertLeft(my_tree[1], 20)
insertRight(my_tree[1], 40)

print("Дерево до видалення гілки:")
print(my_tree)

# Видаляємо гілку з коренем 30 (вузли 30, 20 та 40 зникнуть)
deleteBranch(my_tree, 30)

print("\nДерево після видалення гілки 30:")
print(my_tree)
```

### **Завдання 2:** Оцінити асимптотичну складність (в середньому і в найгіршому випадку) процедур search, insert і delete роботи з деревом.

Search: У середньому: $O(log n)$. У найгіршому разі: $O(n)$. Якщо дерево збалансоване, на кожному кроці пошуку ми відсікаємо половину гілок. У найгіршому разі дерево вироджується в лінійний список, і для пошуку елемента нам доведеться переглянути всі \(n\) елементів один за одним.

Insert: У середньому: $O(log n)$. У найгіршому разі: $O(n)$. Щоб вставити новий елемент, спочатку потрібно знайти для нього правильне пусте місце. Час пошуку цього місця повністю залежить від висоти дерева — від логарифмічної у збалансованому дереві до лінійної у виродженому.

Delete: У середньому: $O(log n)$. У найгіршому разі: $O(n)$. Процес складається з двох етапів: спочатку ми шукаємо потрібний вузол у дереві, а потім перебудовуємо зв'язки (нащадків). Етап пошуку є найдовшим і визначає загальну складність операції.

### **Завдання 4:** Оцінити асимптотичну складність (в середньому і в найгіршому випадку) процедур search, insert і delete роботи з купою.

Search: У середньому: $O(n)$. У найгіршому разі: $O(n)$. Купа в масиві не збалансована по горизонталі, тому швидкого бінарного пошуку немає. Для знаходження довільного елемента нам доведеться послідовно перебрати весь масив один за одним.

Insert: У середньому: $O(1)$. У найгіршому разі: $O(log n)$. Новий елемент записується в кінець масиву і циклом піднімається вгору. У середньому він майже відразу знаходить місце, але в найгіршому разі (якщо він найбільший) йому доведеться піднятися на саму вершину купи.

Delete: У середньому: $O(log n)$. У найгіршому разі: $O(log n)$. Ми переносимо останній елемент на місце видаленого кореня, після чого функція maxHeapify рекурсивно опускає його вниз. Час виконання прямо пропорційний висоті дерева.

### **Завдання 5:** Реалізуйте геш-таблицю з ланцюжковим гешуванням. Проведіть тестування геш-таблиці з різними типами даних (цілі числа, рядки, списки, словники, об’єкти). Перевірити працездатність геш-таблиці з різними типами даних. Виміряти час виконання основних операцій (пошук, вставка, видалення) для різних типів даних.  Порівняти результати для різних типів даних. 

```python
import time

class HashTable:
    def __init__(self):
        self.size = 5
        self.table = [[] for _ in range(self.size)]

    def get_hash(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        slot = self.get_hash(key)
        for pair in self.table[slot]:
            if pair[0] == key:
                pair[1] = value
                return
        self.table[slot].append([key, value])

    def search(self, key):
        slot = self.get_hash(key)
        for pair in self.table[slot]:
            if pair[0] == key:
                return pair[1]
        return None

    def delete(self, key):
        slot = self.get_hash(key)
        for i, pair in enumerate(self.table[slot]):
            if pair[0] == key:
                self.table[slot].pop(i)
                return True
        return False


ht = HashTable()

tests = {
    "Ціле число": 42,
    "Рядок": "привіт",
    "Список": "[1, 2, 3]",
    "Словник": "{'id': 1}",
    "Об'єкт": "Object_User"
}

print("=== ТЕСТУВАННЯ ТА ЗАМІР ЧАСУ ===\n")

for type_name, key in tests.items():
    print(f"--- Тип: {type_name} ---")
    
    start = time.perf_counter()
    ht.insert(key, "тестове_значення")
    print(f"Вставка: успішно за {time.perf_counter() - start:.8f} сек.")
    
    start = time.perf_counter()
    res = ht.search(key)
    print(f"Пошук: знайдено '{res}' за {time.perf_counter() - start:.8f} сек.")
    
    start = time.perf_counter()
    status = ht.delete(key)
    print(f"Видалення: успішно ({status}) за {time.perf_counter() - start:.8f} сек.\n")
```

**Коротке порівняння результатів:** Оскільки всі ключі обробляються через єдину стандартну функцію hash(), час виконання операцій для всіх типів даних є приблизно однаковим і прагне до $O(1)$. Невеликі коливання у часі виникають через випадкове навантаження на процесор комп'ютера під час конкретного тесту, а не через саму структуру даних.
