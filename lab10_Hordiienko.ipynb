Написати процедуру на Python, яка для вхідного повідомлення обчислює список символів chars та список їх частот freq. Опрацювати самостійно тему декодування Гафмена, і скориставшись наведеним там кодом, декодувати повідомлення, задане варіантом. 

```python
import heapq

class Node:
    def __init__(self, freq, symbol, left=None, right=None):
        self.freq = freq
        self.symbol = symbol
        self.left = left
        self.right = right
        self.huff = ''

    def __lt__(self, other):
        return self.freq < other.freq


# Обчислення chars та freq
def get_frequencies(text):
    freq_dict = {}

    for char in text:
        freq_dict[char] = freq_dict.get(char, 0) + 1

    chars = list(freq_dict.keys())
    freq = list(freq_dict.values())

    return chars, freq


# Виведення кодів Гафмена
def print_codes(node, code=''):
    new_code = code + str(node.huff)

    if node.left:
        print_codes(node.left, new_code)

    if node.right:
        print_codes(node.right, new_code)

    if not node.left and not node.right:
        print(f"{node.symbol}: {new_code}")


# Отримання словника кодів
def get_huffman_dict(node, code='', codes=None):
    if codes is None:
        codes = {}

    new_code = code + str(node.huff)

    if node.left:
        get_huffman_dict(node.left, new_code, codes)

    if node.right:
        get_huffman_dict(node.right, new_code, codes)

    if not node.left and not node.right:
        codes[node.symbol] = new_code

    return codes


# Візуалізація дерева
def print_tree(node, level=0):
    if node is not None:
        print_tree(node.right, level + 1)

        print("    " * level + f"{node.symbol}:{node.freq}")

        print_tree(node.left, level + 1)


# Декодування
def decode_huffman(root, encoded_text):
    result = ""
    current = root

    for bit in encoded_text:

        if bit == '0':
            current = current.left
        else:
            current = current.right

        if not current.left and not current.right:
            result += current.symbol
            current = root

    return result


# Повідомлення за варіантом
my_text = "AABAABACBBCBCEEFFFGGGGGGDEE"

# chars і freq
chars, freq = get_frequencies(my_text)

print("chars =", chars)
print("freq =", freq)

# Побудова дерева
nodes = []

for i in range(len(chars)):
    heapq.heappush(nodes, Node(freq[i], chars[i]))

while len(nodes) > 1:

    left = heapq.heappop(nodes)
    right = heapq.heappop(nodes)

    left.huff = 0
    right.huff = 1

    new_node = Node(
        left.freq + right.freq,
        left.symbol + right.symbol,
        left,
        right
    )

    heapq.heappush(nodes, new_node)

root = nodes[0]

print("\nДерево Гафмена:")
print_tree(root)

print("\nКоди Гафмена:")
print_codes(root)

# Отримання кодів
huffman_codes = get_huffman_dict(root)

# Кодування
encoded_text = ""

for char in my_text:
    encoded_text += huffman_codes[char]

print("\nЗакодоване повідомлення:")
print(encoded_text)

# Декодування
decoded_text = decode_huffman(root, encoded_text)

print("\nДекодоване повідомлення:")
print(decoded_text)
```

1. Що таке жадібні алгоритми?

Жадібні алгоритми — це стратегія розв'язання задач, де на кожному кроці обирається локально оптимальне рішення (найкраще «тут і зараз») з надією, що воно приведе до глобального оптимуму.

2. Що таке префіксний код? Який код використовується у коді Гафмена?

Префіксний код — це система кодування, у якій жодне кодове слово не є початком (префіксом) іншого кодового слова. Це дозволяє однозначно декодувати послідовність без роздільників. Код Гафмена є класичним прикладом префіксного коду, де символам, що зустрічаються частіше, призначаються коротші коди.

3. Як пов’язана структура даних «купа» зі структурою даних «черга з пріоритетами»?

Структура даних «купа» (heap) є найефективнішою основою для реалізації «черги з пріоритетами», оскільки вона дозволяє швидко додавати елементи та отримувати/видаляти елемент із найвищим пріоритетом (мінімальним або максимальним значенням) за логарифмічний час $O(\log n)$.

4. Що таке стиснення даних і для чого воно використовується? Які його основні переваги?

Стиснення даних — це процес зменшення обсягу цифрових даних за рахунок усунення надмірності. Воно використовується для економії місця на носіях та прискорення передачі інформації мережами. Головні переваги: зменшення витрат на зберігання, швидкість обміну даними та зниження вимог до пропускної здатності каналів.

5. Які кроки необхідно виконати для стиснення даних за допомогою алгоритму кодування Гафмена?

Для стиснення за алгоритмом Гафмена необхідно: 1) підрахувати частоту появи кожного символу; 2) створити листки дерева для кожного символу; 3) з'єднувати два вузли з найменшими частотами в новий вузол (сума частот), поки не утвориться єдине дерево; 4) призначити коди (наприклад, 0 для лівого відгалуження, 1 для правого) та закодувати текст.

Обмеження алгоритму Гафмена полягають у тому, що він потребує знання частот символів заздалегідь (або передачі таблиці кодів разом із даними) і не враховує залежність між символами (лише їх частоту).
