### **Завдання 1:** Повторити код, наведений в методичці для довільних рядків.

```python
# --- 1. Блок наївного пошуку підрядка ---
def naive_match(pattern, text):
    occurrences = []
    n, m = len(text), len(pattern)
    for i in range(n - m + 1):
        if text[i:i+m] == pattern:
            occurrences.append(i)
    return occurrences

# Тест для пошуку
test_text = "абабагаламага"
test_pattern = "аба"
print(f"--- Тест наївного пошуку ---")
print(f"Текст: '{test_text}', Шаблон: '{test_pattern}'")
print(f"Знайдено на позиціях: {naive_match(test_pattern, test_text)}\n")


# --- 2. Блок Z-функції ---
def z_func(s):
    n = len(s)
    z = [0] * n
    l, r = 0, 0
    for i in range(1, n):
        if i <= r:
            z[i] = min(r - i + 1, z[i - l])
        while i + z[i] < n and s[z[i]] == s[i + z[i]]:
            z[i] += 1
        if i + z[i] - 1 > r:
            l, r = i, i + z[i] - 1
    z[0] = n
    return z

# Тест для Z-функції
test_str = "abacaba"
print(f"--- Тест Z-функції ---")
print(f"Рядок: '{test_str}'")
print(f"Результат Z-функції: {z_func(test_str)}\n")


# --- 3. Блок стиснення через Z-функцію ---
def compress_with_z(s):
    z = z_func(s)
    n = len(s)
    for i in range(1, n):
        # Якщо залишок рядка дорівнює Z-значенню і ділиться на період
        if i + z[i] == n and n % i == 0:
            return s[:i]
    return s

# Тест стиснення
test_comp = "abababab"
print(f"--- Тест стиснення рядка ---")
print(f"Оригінал: '{test_comp}'")
print(f"Стиснутий: '{compress_with_z(test_comp)}'")
```

### **Завдання 2:** Пояснити асимптотику наївного алгоритму пошуку підрядка.

У найгіршому випадку складність становить $O(|t| \cdot |p|)$, де $|t|$ — довжина тексту, $|p|$ — довжина шаблону. Це зумовлено тим, що для кожної позиції в тексті необхідно перевірити всі символи шаблону на збіг.

### **Завдання 3:** Обчислити асимптотичну складність алгоритму стиснення рядка за допомогою 𝑧-функції. 

Обчислення масиву z-функції виконується за $O(n)$ часу, де $n$ — довжина рядка. Подальший прохід по цьому масиву для знаходження стисненого подання також займає $O(n)$ часу. Загальна асимптотична складність алгоритму становить $O(n)$.

### **Контрольні питання:**
1. Що таке «префікс-функція» у контексті алгоритмів на рядках, і як вона відрізняється від Z-функції? 

Префікс-функція $\pi[i]$ для рядка $s$ — це довжина найдовшого власного префікса підрядка $s[0...i]$, який одночасно є його суфіксом. Головна відмінність: префікс-функція фокусується на «накладанні» початку рядка на його кінець у кожній точці, тоді як Z-функція показує довжину найдовшого спільного префікса всього рядка $s$ та його суфікса, що починається з позиції $i$.

2. Що таке Z-функція у вищому контексті алгоритмів на рядках, і яка її роль у вирішенні задач? 

Z-функція — це масив $Z$, де $Z[i]$ дорівнює довжині найдовшого підрядка, який починається в позиції $i$ і збігається з префіксом рядка. Її роль полягає в надшвидкому ($O(n)$) пошуку входжень шаблону в текст, оскільки вона дозволяє виявити всі збіги префіксів за один прохід без необхідності переглядати кожен символ окремо, що є критичним для задач аналізу структури рядків.

3. Які існують підходи до вирішення задачі «найдовший спільний підрядок» для двох рядків? 

Основні методи включають динамічне програмування (матриця збігів з часовою складністю $O(n \cdot m)$), використання суфіксних дерев або суфіксних автоматів (що дозволяють знайти відповідь за $O(n+m)$ часу) та метод бінарного пошуку в поєднанні з хешуванням рядків (для скорочення часу перевірки).

4. Як можна застосувати алгоритми на рядках у задачах обробки природної мови або обробки текстів?

Алгоритми на рядках є фундаментом для NLP: вони використовуються для токенізації (виділення слів), перевірки правопису (виявлення помилок за відстанню Левенштейна), пошуку закономірностей (регулярні вирази), стиснення даних (алгоритм Гафмена) та в задачах машинного перекладу для зіставлення фраз і виявлення семантичної схожості між великими обсягами текстових документів.
