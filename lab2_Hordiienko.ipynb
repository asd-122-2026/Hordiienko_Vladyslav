### **Контрольні питання:**
1. Чому не рекомендується використовувати системний Python у Windows?

Щоб не пошкодити роботу операційної системи та уникнути конфліктів, коли ваші проєкти вимагають різних версій бібліотек, ніж ті, що потрібні самій Windows.

2. Навіщо створювати окреме середовище для кожного проєкту?

Для ізоляції. Це дозволяє встановлювати різні версії пакетів для різних проєктів, щоб вони не заважали один одному.

3. У чому різниця між базовим середовищем і проєктним?

base — це основне середовище Anaconda (його краще залишати чистим), а проєктне — це окрема «кімната» під конкретну роботу, яку можна легко видалити або змінити.

4. Для чого використовується environment.yml?

Він дозволяє швидко відтворити всі необхідні бібліотеки та їх версії на іншому комп'ютері.

5. Як перевірити, який Python використовується в ноутбуці?

Виконати команду where python
