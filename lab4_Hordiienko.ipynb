### **Завдання 1:** Реалізувати процедуру обчислення факторіалу за допомогою циклу for. Оцінити асимптотичну складність алгоритму.


```python
def factorial_for(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result
print (factorial_for(5))
```

Асимптотична складність: $O(n)$, оскільки цикл виконується рівно $n-1$ раз. Кількість операцій множення зростає лінійно залежно від вхідного числа $n$.

### **Завдання 2:** Оцінити асимптотичну складність рекурсивного алгоритму обчислення факторіалу.

```python
def factorial_recursive(n):
    if n <= 1:
        return 1
    return n * factorial_recursive(n - 1)
print (factorial_recursive(5)) 
```

Асимптотична складність: $O(n)$, оскільки функція викликає саму себе $n$ разів. Кожен виклик виконує лише одну операцію множення за сталий час.

### **Завдання 3:** Оцінити асимптотичну складність рекурсивного алгоритму обчислення n-го числа Фібоначчі.

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n - 1) + fib(n - 2)
print (fib(5))
```

Асимптотична складність: $O(2^n)$, оскільки кожен виклик породжує два нових виклики.

### **Завдання 4:** Оцінити асимптотичну складність алгоритму сортування злиттям, використовуючи головну теорему рекурсії.

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    left_sorted = merge_sort(left_half)
    right_sorted = merge_sort(right_half)

    return merge(left_sorted, right_sorted)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

data = [38, 27, 43, 3, 9, 82, 10]
print(f"Оригінал: {data}")
print(f"Результат: {merge_sort(data)}")
```

Асимптотична складність: $O(nlogn)$, оскільки масив розділяється на логарифмічну кількість рівнів ($\log n$), на кожному з яких виконує лінійну роботу ($n$) зі злиття елементів.
