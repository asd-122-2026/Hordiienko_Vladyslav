### **Завдання 1:** Написати функцію pop_n(), що видаляє елементи стека з його початку до номера n включно;


```python
def pop_n(stack, n):
    count = 0
    
    while stack and count < n:
        stack.pop()  
        count += 1
    
    print(f"--- Операція pop_n завершена. Видалено елементів: {count} ---")

my_stack = [10, 20, 30, 40, 50]

print(f"Початковий стек: {my_stack}")
print(f"Вершина стека: {my_stack[-1]}")
print(f"Розмір до видалення: {len(my_stack)}")

pop_n(my_stack, 3)

print(f"Стек після видалення: {my_stack}")
if my_stack:
    print(f"Нова вершина: {my_stack[-1]}")
    print(f"Новий розмір: {len(my_stack)}")
else:
    print("Стек тепер порожній")
```

### **Завдання 3:** Розглянути самостійно Приклад 3 офіційної документації модуля Queue [31];

```python
import queue
import threading
import time

q = queue.Queue()

def worker():
    while True:
        item = q.get()
        print(f"Оброблено: {item}")
        time.sleep(1)

threading.Thread(target=worker, daemon=True).start()

for i in range(5):
    q.put(i)
    time.sleep(0.5)

print("Готово.")
```

### **Завдання 4:** Написати функцію print_n(), що друкує елементи черги з його початку до номера n включно;

```python
from collections import deque

def print_n(queue, n):

    if not queue:
        print("Черга порожня")
        return

    print(f"Перші {n} елементів:")
    
    for i, element in enumerate(queue):
        if i < n:
            print(f"{i + 1}: {element}")
        else:
            break

my_queue = deque(['Яблуко', 'Банан', 'Вишня', 'Драгонфрут', 'Екзотик'])

print_n(my_queue, 3)
```

### **Контрольні питання:**
1. Що таке стек і які операції можна виконувати зі стеком?

Стек — це структура даних, що працює за принципом LIFO («останнім прийшов — першим пішов»). Основні операції: push (додати на вершину), pop (видалити з вершини), peek (переглянути вершину) та isEmpty.
   
2. Яка основна відмінність між стеком та чергою?

Стек працює за принципом LIFO (як стопка тарілок), а черга — за FIFO («першим прийшов — першим пішов», як черга в касу).

3. Як ви можете реалізувати стек за допомогою масиву і за допомогою зв’язаного списку? Які переваги та недоліки кожного підходу?

Масив: швидкий доступ до пам'яті, але має фіксований розмір або потребує дорогої релокації при переповненні.

Зв’язаний список: необмежений розмір (динамічність) та стабільний час вставки $O(1)$, але витрачає більше пам'яті на зберігання вказівників (адрес).

4. Які є застосування стека та черги в програмуванні і реальному житті?

Стек: рекурсія, скасування дій (Undo), перевірка балансу дужок, навігація «назад» у браузері.

Черга: черги на друк, буфери передачі даних, планування завдань ОС та алгоритми обходу графів у ширину (BFS).
